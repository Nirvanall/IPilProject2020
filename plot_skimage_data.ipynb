```
from skimage import data
from skimage import color
import skimage
import matplotlib.pyplot as plt
from skimage.exposure import rescale_intensity
import numpy as np



data1 = skimage.img_as_float(data.binary_blobs(length=128, seed=1))
sigma = 0.35
data1 += np.random.normal(loc=0, scale=sigma, size=data1.shape) # add noise
data1 = rescale_intensity(data1, in_range=(-sigma, 1 + sigma),
                         out_range=(-1, 1))

# to add noise to images: np.random or skimage.util.random_noise()

data2 = color.rgb2gray(data.retina())[300:700, 700:900]

data3 = data.shepp_logan_phantom()



plt.figure()
plt.imshow(data1, cmap = 'gray')
plt.show

plt.figure()
plt.imshow(data2, cmap = 'gray')
plt.show

plt.figure()
plt.imshow(data3, cmap = 'gray')
plt.show
```
